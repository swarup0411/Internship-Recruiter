## Importing the libraries

```python
import pandas as pd
```

## Creating a reference for the skill requirements

```python
web = set([i.strip() for i in 'HTML, CSS, Bootstrap, AJAX, JavaScript, PHP, MySQL, REST API, Deployment, Git, Python, Linux, JSON, JWT'.split(',')])
```

```python
full = set([i.strip() for i in 'HTML, CSS, MongoDB, Express, ReactJS, Node.js, ES6, Redux, JavaScript, Deployment, Python, Linux, JSON, JWT'.split(',')])
```

```python
ml = set([i.strip() for i in 'Python, TensorFlow, PyTorch, NumPy, PyPI, Sci-kit learn, Statistical Modelling, Machine Learning, Deep Learning, SQL, JSON, Deployment, Flask'.split(',')])
```

## Importing the Dataset 

```python
df=pd.read_excel('sample_data_v4.xlsx')
df
```

## Editing for deployment and git 

```python
# The required skills AWS/Heroku/Google Cloud Computing are all clubbed into "Deployment" and Git SVN and Github are clubbed into "Git" because only one othose skills are enough and the company doesn't require all of the skills in the candidate. 
# But it is again restored to the original skills at the end for correct display of output.
 

for x in range(0,5):
    for i in df['Other skills']:
        if type(i)==str:
            if 'Amazon Web Server (AWS)' in i:
                tx = i.replace('Amazon Web Server (AWS)', 'Deployment')
                df.replace(i, tx, inplace=True)
            if 'Amazon Web Services (AWS)' in i:
                tx = i.replace('Amazon Web Services (AWS)', 'Deployment')
                df.replace(i, tx, inplace=True)
            if 'Google Cloud Computing' in i:
                tx = i.replace('Google Cloud Computing', 'Deployment')
                df.replace(i, tx, inplace=True)
            if 'Heroku' in i:
                tx = i.replace('Heroku', 'Deployment')
                df.replace(i, tx, inplace=True)
            if 'Git SVN' in i:
                tx = i.replace('Git SVN', 'Git')
                df.replace(i, tx, inplace=True)
            if 'GitHub' in i:
                tx = i.replace('GitHub', 'Git')
                df.replace(i, tx, inplace=True)
```

## Main Process of selection of candidate 

```python
# The process of selection of a candidate is purely based on hit and trial method. One candidate has to be selected each for 3 roles as an intern in the company.
# So the values pw_limit, pf_limit, pm_limit are set in order to get one candidate each for 3 different roles by hit and trial method. 
# The index, no of skills and probability of getting selected of the winners are stored in the lists winner_web, winner_fullstack, winner_ml. 
# The iteration process goes on for each and every entry in the dataset and the predicted role and skill gap are stored in the lists role_list & skillgap_list respectively. 


role_list=[]
skillgap_list=[]
winner_web, winner_fullstack, winner_ml=[],[],[]
pw_limit, pf_limit, pm_limit = 0.7, 0.55, 0.5
for x in range(0,len(df['Other skills'])):
    i= df['Other skills'].iloc[x]
    if type(i)==str:
        cand_skillset=set([l.strip() for l in i.split(',')])
        web_score,fullstack_score,ml_score=len(cand_skillset&web), len(cand_skillset&full), len(cand_skillset&ml)
        if (web_score/len(web)) > pw_limit:
            if(len(winner_web)==0):
                winner_web.extend([x, len(cand_skillset), web_score/len(web)])
                role_list.append('Web Developer')
                skillgap_list.append('NA')
            else:
                if len(cand_skillset)>winner_web[1]:
                    role_list[winner_web[0]]='NA'
                    s=''
                    for d in web:
                        if d not in cand_skillset:
                            s = s+d+','
                    skillgap_list[winner_web[0]]=s
                    winner_web[0],winner_web[1],winner_web[2]=x,len(cand_skillset),web_score/len(web)
                    role_list.append('Web Developer')
                    skillgap_list.append('NA')
                else:
                    role_list.append('NA')
                    s=''
                    for d in web:
                        if d not in cand_skillset:
                            s = s+d+','
                    skillgap_list.append(s)
                    
        elif (fullstack_score/len(full)) > pf_limit:
            if(len(winner_fullstack)==0):
                winner_fullstack.extend([x, len(cand_skillset), fullstack_score/len(full)])
                role_list.append('FullStack Developer')
                skillgap_list.append('NA')
            else:
                if len(cand_skillset)>winner_fullstack[1]:
                    role_list[winner_fullstack[0]]='NA'
                    s=''
                    for d in full:
                        if d not in cand_skillset:
                            s = s+d+','
                    skillgap_list[winner_fullstack[0]]=s
                    winner_fullstack[0],winner_fullstack[1],winner_fullstack[2]=x,len(cand_skillset),fullstack_score/len(full)
                    role_list.append('FullStack Developer')
                    skillgap_list.append('NA')
                else:
                    role_list.append('NA')
                    s=''
                    for d in full:
                        if d not in cand_skillset:
                            s = s+d+','
                    skillgap_list.append(s)
                    
                    
        elif (ml_score/len(ml)) > pm_limit:
            winner_ml.extend([x,len(cand_skillset), ml_score/len(ml)])
            role_list.append('ML Developer')
            skillgap_list.append('NA')
            
            
        else:
            role_list.append('NA')
            s=''
            if max(web_score, fullstack_score, ml_score)==web_score:
                for d in web:
                    if d not in cand_skillset:
                        s = s+d+','    
            elif max(web_score, fullstack_score, ml_score)==fullstack_score:
                for d in full:
                    if d not in cand_skillset:
                        s = s+d+','
            elif max(web_score, fullstack_score, ml_score)==ml_score:
                for d in ml:
                    if d not in cand_skillset:
                        s = s+d+','
            skillgap_list.append(s)
    else:
        role_list.append('NA')
        skillgap_list.append('NA')

df['Selected for'] = role_list
df['Lacking Skills'] = skillgap_list
```

## Normalisation of the skillset provided by user

```python
# It is the same process as done in the section "Editing for deployment and git" but here, it is done for processing the user input section only


for i in range(len(skill_set)):
    if skill_set[i]=='Amazon Web Server (AWS)' or skill_set[i]=='Amazon Web Serices (AWS)' or skill_set[i]=='Google Cloud Computing' or skill_set[i]=='Heroku' or skill_set[i]=='AWS' or skill_set[i]=='GCP':
        skill_set[i]='Deployment'
    elif skill_set[i]=='Git SVN' or skill_set[i]=='GitHub':
        skill_set[i]= 'Git'
skill_set=set(skill_set)
```

```python
# This section of code only helps in the procssing of user input and displaying the output that the candidate is selected or not and if not selected then his/her skill gap based on his/her desired role and her skillset given in the user input. 
# The shortlisted candidates are already decided and this section of code is only for the info to be given to the user through output whether he/she is selected or not. 
# Here the w2,f2 and m2 are the scores for the skill set given in user input.


if len(skill_set)>0:
    if role == 1:
        w2=len(skill_set&web)
        if(w2/len(web))>=winner_web[2]:
            if len(skill_set)>=winner_web[1]:
                print('You are selected\n')
            else:
                print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
                s=''
                for d in web:
                    if d not in skill_set:
                        s = s+d+','
                if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP')
                print('Skill gap : ',s[:-1])
        else:
            print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
            s=''
            for d in web:
                if d not in skill_set:
                    s = s+d+','
            if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP')
            print('Skill gap : ',s[:-1])
            
    elif role == 2:
        f2=len(skill_set&full)
        if(f2/len(full))>=winner_fullstack[2]:
            if len(skill_set)>=winner_fullstack[1]:
                print('You are selected\n')
            else:
                print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
                s=''
                for d in full:
                    if d not in skill_set:
                        s = s+d+','
                if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP')
                print('Skill gap : ',s[:-1])
        else:
            print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
            s=''
            for d in full:
                if d not in skill_set:
                    s = s+d+','
            if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP')
            print('Skill gap : ',s[:-1])
    elif role == 3:
        m2=len(skill_set&ml)
        if(m2/len(full))>=winner_ml[2]:
            if len(skill_set)>=winner_ml[1]:
                print('You are selected\n')
            else:
                print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
                s=''
                for d in ml:
                    if d not in skill_set:
                        s = s+d+','
                if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP/Heroku')
                print('Skill gap : ',s[:-1])
        else:
            print('You are not selected for your desired role. Thank you for your effort and keep working hard\n')
            s=''
            for d in ml:
                if d not in skill_set:
                    s = s+d+','
            if 'Deployment' in s:
                    s=s.replace('Deployment', 'AWS/GCP/Heroku')
            print('Skill gap : ',s[:-1])
```
